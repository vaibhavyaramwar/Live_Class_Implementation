```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
sns.scatterplot(x=df['sepal_length'],y=df['petal_length'])
plt.show()
```

```python
plt.plot(df['sepal_length'],df['petal_length'],'o')
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
plt.hist(df['sepal_length'])
plt.show()
```

```python
sns.distplot(df['sepal_length'])
plt.show()
```

```python
df1=sns.load_dataset('tips')
```

```python
sns.scatterplot(df1.total_bill,df1.tip)
plt.show()
```

```python
sns.countplot(df1.smoker)
plt.show()
```

```python
df1['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,hue="smoker")
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,style="smoker")
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,style="size")
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,style="time",hue="smoker")
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,size='size')
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df1,kind="line")
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df1,col="time",hue="smoker")
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df1,col="size")
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df1,col="sex",hue="day")
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=df1)
plt.show()
```

```python
sns.catplot(x="day",y="tip",data=df1)
plt.show()
```

```python
sns.catplot(x="day",y="tip",kind="box",hue="sex",data=df1)
plt.show()
```
