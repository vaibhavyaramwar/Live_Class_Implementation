```python
import requests
```

```python
reqResponse = requests.get("https://api.github.com/repos/pandas-dev/pandas/issues")
```

```python
d11 = reqResponse.json()
```

```python
d11[0]['id']
```

```python
for i in range(len(d11)):
    print(d11[i]["id"],d11[i]['url'],d11[i]['repository_url'],d11[i]['comments_url'],sep="||")
    #print(d11[i]['user']['login'],d11[i]['user']['id'])
```

```python
df22=pd.DataFrame(d11,columns=["id",'url','repository_url','comments_url','user'])
```

```python
df22.to_csv("user_data.csv")
```
